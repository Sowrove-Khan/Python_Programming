Matplotlib is very useful for making plot

```python
#Importing matplotlib library

import matplotlib.pyplot as plt
```

```python
#Import numpy to get data for plots
import numpy as np
```

```python
x = np.linspace(0,10,100) #Data between 0 to 10 & display 100 values
y = np.sin(x)             #Displaying the sin angels of 100 values
z = np.cos(x)             #Displaying the cosine angels of 100 values
```

```python
print(x)
print()
print(y)
print()
print(z)
```

Plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()

print()

#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#Adding title x-axis & y-axis labels

plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine Wave')
plt.show()

print() #New line

plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('Cosine Value')
plt.title('Cosine Wave')
plt.show()
```

```python
 #Parabola

 x = np.linspace(-10,10,20)
 y = x**2
 plt.plot(x,y)
 plt.show()
```

```python
#Plot with color and sign

plt.plot(x,y,'r+') #Red color with + symbol
plt.show()
```

```python
#Plot with color and sign

plt.plot(x,y,'g*') #Green color with * symbol
plt.show()
```

```python
#Print multiple line in single plot

a = np.linspace(-5,5,50)
plt.plot(a,np.sin(a),'r--')
plt.plot(a,np.cos(a),'g--')
plt.show()
```

```python
#Creating Bar plot [Very important for analysis]

fig = plt.figure()   #This will create an empty plot
ax = fig.add_axes([0,0,1,1])  #Creating axis 0 means origin 1 means height and width
languages = ['English','French','Spanish','Latin','German']
people = [100,150,35,70,50]
ax.bar(languages,people)     #Bar r x and y axis a ki thakbe seta dilam
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#Creating Pie Chart plot [Very important for analysis]

fig1 = plt.figure()   #This will create an empty plot
ax = fig1.add_axes([0,0,1,1])  #Creating axis 0 means origin 1 means height and width
languages = ['English','French','Spanish','Latin','German']
people = [100,150,35,70,50]
ax.pie(people,labels=languages,autopct='%1.1f%%')     #People ta % dekhabe...autopct function shows how many floating and decimal values shown in figure
plt.show()
```

```python
#Creating Scatter Plot [Very important for clustering]

x = np.linspace(0,10,30) #Taking 0 to 10 range within 30 even values
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])  #Axis add korlam
ax.scatter(x,y,color='g')  #Axis ta scatter korlam,label dilam,color add korlam
ax.scatter(x,z,color='b')
plt.show()
```

```python
#Creating 3D Scatter Plot

fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
