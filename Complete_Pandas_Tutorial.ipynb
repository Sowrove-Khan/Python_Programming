```python
#Importing Diabatics dataset

diabatics_df = pd.read_csv('/content/diabetes.csv')
diabatics_df.head()
```

```python
#Counting the values based on the labels(same value r koita ase seta count kore)
#total kotojon diabatic or not seta easily count kora jai ai way te

diabatics_df.value_counts('Outcome')
```

```python
#Group the values based on the mean value of all columns
# 0 label r modde all column r mean value & 1 label r modde all column r mean value show hobe
#We can easily understand all the mean values of different labels

diabatics_df.groupby('Outcome').mean()
```

Statistical Measures

```python
# Count the number of values in each columns
diabatics_df.count()

#Mean values of all columns
diabatics_df.mean()

#Standard Daviation of all columns
diabatics_df.std()

#Minimum values of all columns
diabatics_df.min()

#Maximum values of all columns
diabatics_df.max()

#Show all values ot the same time
diabatics_df.describe()
```

Manipulating a DataFrame

```python
#Adding a column to a dataframe [Result name new column add kore tar value dilam Outcome r]
diabatics_df['Result'] = diabatics_df.Outcome
diabatics_df.head()
```

```python
#Removing a row from DataFrame [we want to remove row 1 (so index=0 axis=0)]
#Aivabe remove korle parmanantly remove hobe na.Temporary hobe.

diabatics_df.drop(index=0,axis=0)
```

```python
#Removing a column from DataFrame [we want to remove column 9 (so columns='Outcome' axis=1)]
#Aivabe remove korle parmanantly remove hobe na.Temporary hobe.

diabatics_df.drop(columns='Outcome',axis=1)
```

```python
#Display a specific row using the index value

diabatics_df.iloc[1]
```

```python
#Display a specific column

print(diabatics_df.iloc[:,0])   #print the first column
print(diabatics_df.iloc[:,1])   #print the second column
print(diabatics_df.iloc[:,2])   #print the third column
print(diabatics_df.iloc[:,-1])  #print the last column
```

Correlation :

1.Positive Correlation [when one column is increases the other column increases too]

2.Negative Correlation [when one column is increases the other column decreases]

```python
#To display correlation [very important for analysis]

diabatics_df.corr()
```
